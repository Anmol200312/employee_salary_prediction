```python
# Importing ml algorithms
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
```

```python
# Importing label encoder
from sklearn.preprocessing import LabelEncoder
label_cols = ['workclass', 'marital-status', 'occupation', 'relationship', 'race', 'gender', 'native-country']

for col in label_cols:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])

# Feature and target split
x = data.drop(columns=['income'])  # Features
y = data['income']                 # Target

# Train-Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Min-Max Scaling (AFTER label encoding and split)
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Random Forest": RandomForestClassifier(),
    "SVM (LinearSVC)": LinearSVC(max_iter=10000),
    "Gradient Boosting": GradientBoostingClassifier(),
    "K-Nearest Neighbors": KNeighborsClassifier(n_neighbors=5),  # you can adjust n_neighbors
    "MLP (Neural Net)": MLPClassifier(hidden_layer_sizes=(100,), max_iter=500)
}
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

accuracies = []
model_names = []
trained_models = []

for name, model in models.items():
    print(f"\n{name}")
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    acc = accuracy_score(y_test, y_pred)
    accuracies.append(acc)
    model_names.append(name)
    trained_models.append(model)

    print("Accuracy:", acc)
    print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
    print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
plt.figure(figsize=(10, 6))
bars = plt.barh(model_names, accuracies, color='mediumseagreen')
plt.xlabel("Accuracy")
plt.title("Accuracy Comparison of ML Models")
plt.xlim(0, 1)

# Add value labels to each bar
for bar, acc in zip(bars, accuracies):
    plt.text(acc + 0.01, bar.get_y() + bar.get_height()/2, f"{acc:.2f}", va='center')

plt.gca().invert_yaxis() 
plt.tight_layout()
plt.show()
```

```python
best_index = accuracies.index(max(accuracies))
best_model_name = model_names[best_index]
best_accuracy = accuracies[best_index]
best_model = trained_models[best_index] 


print(f"\n Best Model: {best_model_name} with Accuracy: {best_accuracy:.4f}")

# Save the best model using joblib
import joblib
joblib.dump(best_model, f"{best_model_name}_best_model.pkl")

print(f"\n Best Model '{best_model_name}' saved as '{best_model_name}_best_model.pkl'")
```
